```python
import pandas as pd

clientes = pd.read_csv('data/olist_customers_dataset.csv')
pagamento = pd.read_csv('data/olist_order_payments_dataset.csv')
avaliacao = pd.read_csv('data/olist_order_reviews_dataset.csv')
localizacao = pd.read_csv('data/olist_geolocation_dataset.csv')  
pedidos = pd.read_csv('data/olist_orders_dataset.csv')  
produtos = pd.read_csv('data/olist_products_dataset.csv')  
vendedores = pd.read_csv('data/olist_sellers_dataset.csv')  
categoria_produto = pd.read_csv('data/product_category_name_translation.csv')
itens_pedido = pd.read_csv('data/olist_order_items_dataset.csv')
```

```python
itens_pedido
# converter shipping_limit_date para apenas data
itens_pedido["shipping_limit_date"] = (
    pd.to_datetime(itens_pedido["shipping_limit_date"]).dt.date
)
itens_pedido.to_csv('trusted/itens_pedido.csv')
```

```python
vendedores.to_csv('trusted/vendedores.csv')
print('Vendedores Salvos!')
produtos.to_csv('trusted/produtos.csv')
print('produtos salvos')
categoria_produto.to_csv('trusted/categoria_produto.csv')
print('categoria salva')
```

```python
# apenas hora
pedidos["order_approved_at"] = (
    pd.to_datetime(pedidos["order_approved_at"]).dt.time
)


# apenas datas 
cols = [
    "order_delivered_carrier_date",
    "order_delivered_customer_date",
    "order_estimated_delivery_date",
    "order_purchase_timestamp",
]

pedidos[cols] = pedidos[cols].apply(
    lambda c: pd.to_datetime(c, errors="coerce").dt.date
)

pedidos.to_csv('trusted/pedidos.csv')
print('pedidos salvos')
```

```python
localizacao.to_csv('trusted/localização.csv')
```

```python
quantidade_nulos = avaliacao.isnull
quantidade_nulos
avaliacao_renomeada = avaliacao.fillna('Não avaliou')
avaliacao_renomeada
avaliacao_renomeada.to_csv('trusted/avaliacao.csv')
print('avaliação salva em trusted')
```

```python
quantidade_nulos = pagamento.isnull().sum()
print(quantidade_nulos)
# Certifique-se de converter para string antes de usar o .str.replace
pagamento['payment_value'] = pagamento['payment_value'].astype(str).str.replace('.', ',', regex=False)
pagamento_limpo = pagamento.drop('payment value', axis=1)
# salvando como .csv no diretorio trusted
clientes.to_csv('trusted/pagamento.csv')


print('pagamentos alvos como csv!')
```
